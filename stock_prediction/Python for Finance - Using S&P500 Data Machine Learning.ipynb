```python
import numpy as np
import pandas as pd
import pickle
```

```python
def process_ticker_data(ticker):
    hm_days = 5
    df = pd.read_csv('sp500_joined_closes_2000.csv', index_col=0)
    global tickers
    tickers = df.columns.values.tolist()
    df.fillna(0, inplace=True)
    
    #calculate the percentage future change and for i days
    for i in range(1, hm_days+1):
        df['{}_{}d'.format(ticker, i)] = (df[ticker].shift(-i) - df[ticker]) / df[ticker]
    df.fillna(0, inplace=True)
    return df

process_ticker_data('FB')
```

```python
from collections import Counter

def buy_sell_hold(*args):
    #for each of the next 5 days, if there is a more than 2% increase, gives buy signal
    cols = [c for c in args]
    requirement = 0.02
    for col in cols:
        if col > requirement:
            return 1   #buy
        if col < -requirement:
            return -1  #sell
    return 0 #hold

def extract_futuresets(ticker):
    df = process_ticker_data(ticker)
    df['{}_target'.format(ticker)] = list(map(buy_sell_hold, 
                                              df['{}_1d'.format(ticker)], 
                                              df['{}_2d'.format(ticker)], 
                                              df['{}_3d'.format(ticker)],
                                              df['{}_4d'.format(ticker)],
                                              df['{}_5d'.format(ticker)]))
    vals = df['{}_target'.format(ticker)].values.tolist()
    str_vals = [str(i) for i in vals]
    #we want to get a perfect spread as stock
    print('Data spread: ', Counter(str_vals), sep='')
    
    #cleaning up data
    #replace all infinities as there are some 0 prices at some particular dates for newer stock
    df.fillna(0, inplace=True)
    df = df.replace([np.inf, -np.inf], np.nan)
    #drop all rows with incomplete data
    df.dropna(inplace=True)
    
    #normalizing training data
    #replace all infinities as there are some 0 prices at some particular dates for newer stock
    df_vals = df[[ticker for ticker in tickers]].pct_change()
    df_vals = df_vals.replace([np.inf, -np.inf], 0)
    df_vals.fillna(0, inplace=True)
    
    #getting labeled data X, Y
    X = df_vals.values #only values are returned as a numpy array
    Y = df['{}_target'.format(ticker)].values
    
    return X, Y, df
    
extract_futuresets('FB')
```

```python
from sklearn import svm, model_selection, neighbors
from sklearn.ensemble import VotingClassifier, RandomForestClassifier

def do_ml(ticker):
    X, Y, df = extract_futuresets(ticker)
    #splitting 20% of the data as test data  
    X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size = 0.2)
    
    #using different types of classifiers
    #clf = neighbors.KNeighborsClassifier()
    clf = VotingClassifier([('lsvc', svm.LinearSVC()),
                            ('knn', neighbors.KNeighborsClassifier()),
                            ('rfor', RandomForestClassifier())])
    
    clf.fit(X_train, Y_train)
    confidence = clf.score(X_test, Y_test)
    print('Accuracy', confidence)
    #saving models
    pickle.dump(clf, open('{}_model.pickle'.format(ticker), 'wb'))
    
    #if you are satisfied with the score
    #you can pickle out the model and load then predict
    prediction = clf.predict(X_test)
    print(prediction)
    #print(X_test)
    print('Predicted spread:', Counter(prediction))
    
    return

do_ml('FB')    
```
